### Some Reminders on Exploratory data analysis (EDA)

Exploratory data analysis (EDA) is an especially important activity in the routine of a data analyst or scientist. It enables an in depth understanding of the dataset, define or discard hypotheses and create predictive models on a solid basis. It uses data manipulation techniques and several statistical tools to describe and understand the relationship between variables and how these can impact business. By means of EDA, we can obtain meaningful insights that can impact analysis under the following questions (If a checklist is good enough for pilots to use every flight, it’s good enough for data scientists to use with every dataset).
1. What question are you trying to solve (or prove wrong)?
2. What kind of data do you have?
3. What’s missing from the data?
4. Where are the outliers?
5. How can you add, change or remove features to get more out of your data?

**``Exploratory data analysis (EDA)``** is often an **iterative brainstorming process** where you pose a question, review the data, and develop further questions to investigate before beginning model development work.
```python
# 28 fuel_consumption : All filled data is a list contained two items are list

df.fuel_consumption = df.fuel_consumption.apply(lambda x: ' '.join([i  for list in x for i in list]) if type(x) == list else x)

df.fuel_consumption.str.contains('/100').sum() == df.fuel_consumption.notnull().sum()    # True
df.fuel_consumption.str.contains('comb').sum() == df.fuel_consumption.notnull().sum()    # False
df.fuel_consumption.str.contains('comb|city|country').sum() == df.fuel_consumption.notnull().sum()
# All fiilled data is contained /100 km (comb., city, country), but not all contains comb

(df.fuel_consumption.str.contains('l/100 km', regex = False).sum() + df.fuel_consumption.str.contains('kg/100 km', regex = False).sum()) ==  df.fuel_consumption.notnull().sum()
# All fiilled data is contained either l/100 km or kg/100 km

df.fuel_consumption = df.fuel_consumption.apply(lambda x: x.split('/100 km')[0] if type(x) == str else x).str.strip()

# All kg/100 km turn into l/100 
fuel_consumption_factor = {
'Benzine' : 1.35,
'Biogas':0.5,
'Diesel': 0.832,
'CNG' :  0.74,
'Hydrogen': 2.75,
'LPG': 0.51}

def fuel_consumption_calc(x):
    if isinstance(x['fuel_consumption'], str) and x['fuel_consumption'].endswith('kg'):
        if isinstance(x['fuel_type'], str):
            return round(float(x['fuel_consumption'][:-2]) * fuel_consumption_factor[x['fuel_type']], 2)

    elif isinstance(x['fuel_consumption'], str) and x['fuel_consumption'].endswith('l'):
        return float(x['fuel_consumption'][:-2])

    else:
        return x['fuel_consumption']      
   
df.fuel_consumption = df[['fuel_consumption', 'fuel_type']].apply(fuel_consumption_calc, axis = 1)
```

```python
# showing droppimg features

def general_information():
    print('#\t Column\t\t\tNon-Null Count (%)\t NAN Count (% NAN)\t Dtype\t\tUnique(Category)')
    print('-'*115)   
    for i in range(len(df.columns)):
        text= ''
        print(f'{i}\t{df.columns[i]:<25}', end = '')
        print(f'{df.iloc[:,i].notnull().sum():<7} {(df.iloc[:,i].notnull().sum()/len(df)*100).round(2):<17}', end = '')
        
        nan_per = (df.iloc[:,i].isnull().sum()/len(df)*100).round(2)
        if nan_per > 70:
            text = 'CLEAR'   
        
        print(f'{df.iloc[:,i].isnull().sum():<7} {(df.iloc[:,i].isnull().sum()/len(df)*100).round(2):<4} {text:<10}', end = '')    
        print(f'{str(df.iloc[:,i].dtype):<20}', end = '')
        print(len(df.iloc[:,i].unique()) if (df.iloc[:,i].dtype) == 'category' else'-')   

general_information()
```
