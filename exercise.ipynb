exercise
A group of people is taken and put into an experiment. The aim of this experiment is to measure the pulse rate of people during their actions, depending on whether they consume fat while dieting.

· id -> Number of person being tested (1–30)

· diet -> How much fat is fed while dieting(no fat,low fat)

· pulse -> Person’s heart rate(80–150)

· time -> Time elapsed during his/her action

· kind -> His/Her situation while being tested

```python
df.columns = ['Sıra No', 'Denek Mahlas', 'Yemek Alışkanlığı', 'Kalp Atımı','Yapılan Süre','Yapılan Faaliyet']
#df.rename = ({'Unnamed': 'Sıra No', 'id': 'Denek Mahlas', 'diet': 'Yemek Alışkanlığı', 'pulse' : 'Kalp Atımı','time': 'Yapılan Süre','kind': 'Yapılan'}, axis = 1 inplace = True)
#df.rename = dict(Unnamed = 'Sıra No', id =  'Denek Mahlas', diet = 'Yemek Alışkanlığı', pulse = 'Kalp Atımı', time = 'Yapılan Süre', kind = 'Yapılan Faaliyet')

#df = df.rename(index={'Unnamed': 'Sıra No', 'id': 'Denek Mahlas', 'diet': 'Yemek Alışkanlığı', 'pulse' : 'Kalp Atımı','time': 'Yapılan Süre','kind': 'Yapılan'}, inplace=True)

df.sample(10)
```

```python
df.describe(include = 'all')
```

```python
show = df.describe(include = 'all')
show.rename(index = {'count':'observation'}, inplace=True)

null_counts = df.isnull().sum()
show.loc['null counts'] = null_counts

show.loc[['observation', 'null counts']]
```

```python
summarize = df.describe()

numerical_feature = df.select_dtypes(include =['int64'])
summarize = numerical_feature.describe()

summarize.rename(index={'count':'observation', '50%': 'median'}, inplace=True)

modes = df.mode()
summarize.loc['mode'] = modes.loc[0]

print('Central Tendency')
summarize.loc[['observation', 'mean', 'median','mode']]
```

```python
sns.histplot(x = 'pulse', data = df) ;
```

```python
import scipy as sp

ranges = numerical_feature.max() - numerical_feature.min()
summarize.loc['range'] = ranges

iqrs = sp.stats.iqr(numerical_feature)
summarize.loc['iqr'] = iqrs

summarize.rename(index={'std':'standard derivation', '25%': 'Q1','75%':'Q3'}, inplace=True)

print('Dispersion')
summarize.loc[['min','max','range', 'standard derivation', 'Q1', 'Q3','iqr']]
```
