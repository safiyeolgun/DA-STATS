```python
sales["date"] = pd.to_datetime(sales["date"]).dt.date

sales["date"] = sales["date"].apply(lambda x : x.strftime('%d/%m/%Y'))
sales
```

```python
sales.info()
```

```python
rates["date"] = pd.to_datetime(rates["date"]).dt.date

rates["date"] = rates["date"].apply(lambda x : x.strftime('%d/%m/%Y'))
rates
```

```python
merged_df = pd.merge(sales, rates, on = ["currency", "date"])
merged_df
```

```python
merged_df.isnull().sum()
```

```python
merged_df["total_fiyat"] = merged_df["price"] / merged_df["rate"]
merged_df
```

```python
merged_df["total_fiyat"].sum()
```
